## Differentiation
우선 미분 알고리즘에 대해 알아보자.

```
import numpy as np
import matplotlib.pyplot as plt
```

씹고 뜯고 맛보고 즐길 함수 `f`를 정의한다.

```
def f(x):
    return x + np.cos(x)
```

함수를 구경해 보자.

```
plt.figure()
plt.title("Function Plot")
plt.plot(np.linspace(0, 10, 1000), f(np.linspace(0, 10, 1000)))
plt.show()
```

### Forward & Central Difference

Forward difference와 central difference를 계산해 보자.

```
def central_diff(f, x):
    steps = len(x) - 1
    h = len(x) / steps
    result = np.empty((steps + 1,))
    for i in range(steps):
        result[i] = (f(x[i] + h / 2) - f(x[i] - h / 2)) / h
    result[-1] = result[-2]
    return result

def forward_diff(f, x):
    steps = len(x) - 1
    h = len(x) / steps
    result = np.empty((steps + 1,))
    for i in range(steps):
        result[i] = (f(x[i] + h) - f(x[i])) / h
    result[-1] = result[-2]
    return result
```

앞서 나온 함수 $y=x+\cos x$를 구현한 함수들로 미분한다.

```
x = np.linspace(0, 10, 1000)
diff_forward = forward_diff(f, x)
diff_cent = central_diff(f, x)
```

$y'=1-\sin x$이므로, 실제 도함수와 미분한 결과를 비교하면 central difference가 더 정확한 것을 알 수 있다.

```
plt.figure()
plt.plot(x, diff_forward, 'g-', label='forward')
plt.plot(x, diff_cent, 'b-', label='central')
plt.plot(x, 1-np.sin(x), 'r-', label='real')
plt.legend()
plt.show()
```
